```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from functools import reduce

%matplotlib inline
```

```python
df = pd.read_csv('movies_metadata.csv', low_memory=False)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['imdb_id', 'overview', 'belongs_to_collection', 'original_language', 'id', 'original_title' ,'homepage', 'poster_path', 'production_companies','production_countries','release_date','spoken_languages','status','tagline','title','video'], axis=1, inplace=True)
```

```python
df['vote_average'].fillna(df['vote_average'].mode()[0], inplace=True)
df['vote_count'].fillna(df['vote_count'].mode()[0], inplace=True)
df['popularity'].fillna(df['popularity'].mode()[0], inplace=True)
df['runtime'].fillna(df['runtime'].mode()[0], inplace=True)
df['revenue'].fillna(df['revenue'].mode()[0], inplace=True)
df['budget'] = df['budget'].apply(int)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
data = pd.concat([df['revenue'], df['vote_average']], axis=1)
data.plot.scatter(x='revenue', y='vote_average');
```

```python
data = pd.concat([df['budget'], df['vote_average']], axis=1)
data.plot.scatter(x='budget', y='vote_average');
```

```python
def change_genre(genre):
    if genre.find("Animation") == -1:
        return 'regular'
    else:
        return 'animation'
```

```python
def change_vote_average(vote):
    if vote <= 3.5:
        return 0
    elif 3.5 < vote <= 6.5:
        return 1
    else:
        return 2
```

```python
df['genres'] = df['genres'].apply(change_genre)
df['vote_average'] = df['vote_average'].apply(change_vote_average)
```

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(16, 12))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corrmat, cmap=cmap, vmax=.8, square=True, center=0)
```

```python
#scatterplot
sns.set()
cols = ['vote_average', 'vote_count', 'runtime', 'revenue', 'budget', 'popularity', 'adult']
sns.pairplot(df[cols], size = 2.5)
plt.show();
```

```python
y_train = df['vote_average']
X_train = df.drop(['genres','vote_average'], axis=1)
animation = df['genres'] == 'animation'
y_test = df[animation]['vote_average']
X_test = df[animation].drop(['genres','vote_average'], axis=1)
```

```python
sns.distplot(y_train);
sns.distplot(y_test);
```

```python
# feature importance using random forest
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=80, max_features='auto')
rf.fit(X_train, y_train)
print('Training done using Random Forest')

ranking = np.argsort(-rf.feature_importances_)
f, ax = plt.subplots(figsize=(11, 9))
sns.barplot(x=rf.feature_importances_[ranking], y=X_train.columns.values[ranking], orient='h')
ax.set_xlabel("Feature Importance")
plt.tight_layout()
plt.show()
print(np.sort(rf.feature_importances_))
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

sums = []
n = 4
print(X_train.iloc[:,ranking].columns)
X_train2 = X_train.iloc[:,ranking[:n]]
train_mx = X_train2.values
a = np.sort(rf.feature_importances_[range(n)])
a = a[::-1]
#print('len of test = ' + str(len(train_mx)) + ' len of y = ' + str(len(y_train)) + '\n')
dot_result = train_mx.dot(a)
dot_result = dot_result/np.mean(dot_result)
y = np.exp(y_train)
plt.scatter(x=dot_result,y=y)
plt.scatter(x=y,y=y)
rms1 = sqrt(mean_squared_error(y_train, dot_result))
print(rms1)
```

```python
sc = StandardScaler()
sc.fit(X_train)

X_train_std = sc.transform(X_train)

ppn = Perceptron(max_iter=40, eta0=0.1, random_state=0)

# Train the perceptron
ppn.fit(X_train_std, y_train.values.ravel().astype('int'))

y_pred = ppn.predict(X_train_std)
accuracy1 = accuracy_score(y_train.values.ravel().astype('int'), y_pred)
print('Accuracy: %.2f' % accuracy1)
```

```python
sums = []
print(X_test.iloc[:,ranking].columns)
X_test2 = X_test.iloc[:,ranking[:n]]
test_mx = X_test2.values
a = np.sort(rf.feature_importances_[range(n)])
a = a[::-1]
dot_result = test_mx.dot(a)
dot_result = dot_result/np.mean(dot_result)
y = np.exp(y_test)
plt.scatter(x=dot_result,y=y)
plt.scatter(x=y,y=y)
rms2 = sqrt(mean_squared_error(y_test, dot_result))
print(rms2)
```

```python
X_test_std = sc.transform(X_test)

# Train the perceptron
ppn.fit(X_test_std, y_test.values.ravel().astype('int'))

y_pred = ppn.predict(X_test_std)
accuracy2 = accuracy_score(y_test.values.ravel().astype('int'), y_pred)
print('Accuracy: %.2f' % accuracy2)
```
